```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from scipy.stats import shapiro
```

```python
shapiro(df["Sales"])
```

```python
shapiro(df["Profit"])
```

```python
df["Order Date"] = pd.DatetimeIndex(df["Order Date"])
df.head()
```

```python
upper = df["Profit"].mean() + 3*df["Profit"].std()
lower = df["Profit"].mean() - 3*df["Profit"].std()
upper, lower
```

```python
sns.boxplot(
    data=df,
    x="Profit",
    y="Order Date"    
)
#penampilan outlier
```

```python
newdf = df.copy()
newdf.loc[(newdf["Profit"]>upper), "Profit"] = upper
newdf.loc[(newdf["Profit"]<lower), "Profit"] = lower
```

```python
sns.boxplot(
    data=newdf,
    x="Profit",
    y="Order Date"    
)
#Tanpa Outlier
```

```python
upperS = df["Sales"].mean() + 3*df["Sales"].std()
lowerS = df["Sales"].mean() - 3*df["Sales"].std()
upperS, lowerS
```

```python
df.loc[(df["Sales"] > upperS) | (df["Sales"] < lowerS)]
#jumlah outlier sales
```

```python
sns.lineplot(
    data=newdf,
    x="Order Date",
    y="Sales"    
)
#sales dengan outlier
```

```python
newdf.loc[(newdf["Sales"]>upperS), "Sales"] = upperS
newdf.loc[(newdf["Sales"]<lowerS), "Sales"] = lowerS
```

```python
sns.lineplot(
    data=newdf,
    x="Order Date",
    y="Sales"    
)
#sales tanpa outlier
```

```python
sns.boxplot(
    data=newdf,
    x="Profit",
    y="Order Date"    
)
#Tanpa Outlier
```
